```python
def unique_sorted_letters(text):
    """
    Takes a string and returns a sorted list of unique letters.
    """
    return sorted(set(text))

word = "cheese"
print(f"Original: {word}")
print(f"Unique sorted: {unique_sorted_letters(word)}")
```

```python
def letters_in_either(word1, word2):
    """Returns sorted list of letters in at least one of the two words (Union)."""
    return sorted(set(word1) | set(word2))

def letters_in_both(word1, word2):
    """Returns sorted list of letters in both words (Intersection)."""
    return sorted(set(word1) & set(word2))

def letters_in_one_but_not_both(word1, word2):
    """Returns sorted list of letters in either word, but not in both (Symmetric Difference)."""
    return sorted(set(word1) ^ set(word2))

w1 = "cheese"
w2 = "cake"
print(f"Either: {letters_in_either(w1, w2)}")   
print(f"Both:   {letters_in_both(w1, w2)}")     
print(f"Diff:   {letters_in_one_but_not_both(w1, w2)}") 
```

```python
def capital_city_manager():
    capitals = {
        "france": "Paris",
        "germany": "Berlin",
        "uk": "London"
    }

    print("Type 'quit' to exit.")
    
    while True:
        country = input("Enter a country name: ").strip().lower()
        
        if country == 'quit':
            break
            
        if country in capitals:
            print(f"The capital of {country.title()} is {capitals[country]}.")
        else:
            new_capital = input(f"I don't know the capital of {country.title()}. Please tell me: ")
            capitals[country] = new_capital
            print("Thanks, I've updated my database.")

```

```python
def frequency_analysis(message):
    """
    Reports the six most common letters and their counts.
    Ignores case.
    """

    message = message.lower()
    counts = {}
    
    for char in message:
        if char.isalpha():
            counts[char] = counts.get(char, 0) + 1

    sorted_counts = sorted(counts.items(), key=lambda x: x[1], reverse=True)

    return sorted_counts[:6]

text = "This is a test sentence to see which letters appear most frequently."
print(frequency_analysis(text))
```
