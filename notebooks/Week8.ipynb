```python
import sys

def nl_command(filename):
    try:
        with open(filename, 'r') as file:
            line_number = 1
            for line in file:
                print(f"{line_number}\t{line}", end="")
                line_number += 1
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")

if __name__ == "__main__":
    if len(sys.argv) != 2:
        print("Usage: python nl.py <filename>")
    else:
        nl_command(sys.argv[1])
```

```python
import sys

def diff_command(file1, file2):
    try:
        with open(file1, 'r') as f1, open(file2, 'r') as f2:
            content1 = f1.read()
            content2 = f2.read()
            
            if content1 == content2:
                print("The files are the same.")
            else:
                print("The files are different.")
                
    except FileNotFoundError as e:
        print(f"Error: Could not open one of the files. {e}")

if __name__ == "__main__":
    if len(sys.argv) != 3:
        print("Usage: python diff.py <file1> <file2>")
    else:
        diff_command(sys.argv[1], sys.argv[2])
```

```python
import sys

def grep_command(pattern, filename):
    try:
        with open(filename, 'r') as file:
            for line in file:
                if pattern in line:
                    print(line, end="")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")

if __name__ == "__main__":
    if len(sys.argv) != 3:
        print("Usage: python grep.py <pattern> <filename>")
    else:
        grep_command(sys.argv[1], sys.argv[2])
```

```python
import sys

def wc_command(filename):
    line_count = 0
    char_count = 0
    
    try:
        with open(filename, 'r') as file:
            for line in file:
                line_count += 1
                char_count += len(line)
        
        print(f"Lines: {line_count}")
        print(f"Characters: {char_count}")
        
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")

if __name__ == "__main__":
    if len(sys.argv) != 2:
        print("Usage: python wc.py <filename>")
    else:
        wc_command(sys.argv[1])
```

```python
import sys
import string

def load_dictionary(dict_file):
    valid_words = set()
    try:
        with open(dict_file, 'r') as f:
            for line in f:
                valid_words.add(line.strip().lower())
    except FileNotFoundError:
        print("Error: Dictionary file not found.")
        sys.exit(1)
    return valid_words

def spell_check(filename, valid_words):
    try:
        with open(filename, 'r') as file:
            content = file.read()
        
            translator = str.maketrans(string.punctuation, ' ' * len(string.punctuation))
            clean_content = content.translate(translator)
            
            words = clean_content.split()
            
            unknown_words = set()
            for word in words:

                if word.lower() not in valid_words:
                    unknown_words.add(word)
            
            print("Unknown words found:")
            for word in unknown_words:
                print(word)
                
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")

if __name__ == "__main__":
    if len(sys.argv) != 2:
        print("Usage: python spell.py <filename>")
    else:
        dictionary_path = 'words.txt' 
        
        dictionary_set = load_dictionary(dictionary_path)
        spell_check(sys.argv[1], dictionary_set)
```
