```python
def int_to_binary(n):
    """
    Accepts a positive integer and returns its binary representation.
    """

    return bin(n)[2:]
number = 10
print(f"Binary of {number}: {int_to_binary(number)}") 
```

```python
def get_factors(n):
    """
    Takes an integer and returns a list of its factors.
    """
    factors = []
    for i in range(1, n + 1):
        if n % i == 0:
            factors.append(i)
    return factors

num = 12
print(f"Factors of {num}: {get_factors(num)}") 
```

```python
def is_prime(n):
    """
    Determines if a given integer is a prime number.
    """
    if n <= 1:
        return False
    
    for i in range(2, n):
        if n % i == 0:
            return False 
            
    return True


print(f"Is 7 prime? {is_prime(7)}")   
print(f"Is 10 prime? {is_prime(10)}") 
```

```python
def obfuscate_message(message):
    """
    Removes spaces and reverses the string.
    """
    no_spaces = message.replace(" ", "")
    
    reversed_msg = no_spaces[::-1]
    
    return reversed_msg


msg = "send cheese"
print(f"Original: {msg}")
print(f"Encrypted: {obfuscate_message(msg)}")
```

```python
import random
import string

def encrypt_with_noise(message):
    """
    Encrypts a message by inserting random letters at a random interval.
    Returns the encrypted message and the interval used.
    """
    interval = random.randint(2, 20)
    encrypted_msg = ""
    
    for char in message:
        encrypted_msg += char
        for _ in range(interval - 1):
            encrypted_msg += random.choice(string.ascii_letters)
            
    return encrypted_msg, interval

original = "send cheese"
encrypted_text, key = encrypt_with_noise(original)
print(f"Interval: {key}")
print(f"Encrypted: {encrypted_text}")
```

```python
def decrypt_noise(encrypted_message, interval):
    """
    Decrypts a message by extracting characters at the specified interval.
    """
    return encrypted_message[::interval]

decrypted_text = decrypt_noise(encrypted_text, key)
print(f"Decrypted: {decrypted_text}")
```
