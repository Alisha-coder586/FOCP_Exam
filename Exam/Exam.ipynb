```python
def check_for_coconut(item_list):
    """
    Returns True if the list contains the string 'coconut' (lowercase),
    otherwise returns False.
    """
    if "coconut" in item_list:
        return True
    else:
        return False

# Code to demonstrate the function works
list_with_fruit = ["apple", "banana", "coconut", "date"]
list_without_fruit = ["apple", "banana", "date"]

print(f"List 1 (has coconut): {check_for_coconut(list_with_fruit)}")
print(f"List 2 (no coconut): {check_for_coconut(list_without_fruit)}")
```

```python
user_strings = []

while True:
    # Prompt user for input
    entry = input("Enter a string (or press Enter to finish): ")

    # Check if input is empty (user pressed Enter without typing)
    if entry == "":
        break

    # Add the string to the list
    user_strings.append(entry)

# Process the list: convert to lower case and sort alphabetically
# We use a list comprehension to create a new list of lower case strings
processed_list = [s.lower() for s in user_strings]
processed_list.sort()

# Display the results
print("\nYour strings in alphabetical order:")
for s in processed_list:
    print(s)
```
