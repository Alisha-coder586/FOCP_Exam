```python
import random

def main():
    while True:
        # 1. Prompt for password
        password = input("Enter your password: ")

        # 2. Check length (must be 9 or greater)
        if len(password) < 9:
            print("Password too short.")
            print("Please try again.\n")
            continue # RESTART LOOP

        # 3. Ask for three random letters
        check_failed = False
        
        for _ in range(3):
            # Select a random index
            random_index = random.randint(0, len(password) - 1)
            
            # Display position (human-readable 1-based index)
            user_char = input(f"Enter letter at position {random_index + 1}: ")

            # Check match
            if user_char == password[random_index]:
                print("Correct")
            else:
                print("Security check failed.")
                check_failed = True
                break # Break out of the letter loop

        # 4. Success message or retry
        if not check_failed:
            print("Security check passed.")
            break # EXIT the main loop
        else:
            print("Please try again.\n")

if __name__ == "__main__":
    main()
```
