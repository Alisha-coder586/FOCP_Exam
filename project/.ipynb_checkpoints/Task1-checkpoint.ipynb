```python
def get_valid_price(pizza_number):
    """
    Prompts the user for a pizza price and ensures it is a positive number.
    Repeats the prompt until valid input is received.
    """
    while True:
        try:
            # Prompt user for specific pizza number [cite: 1344]
            user_input = input(f"Enter Price of Pizza #{pizza_number}: ")
            price = float(user_input)
            
            # Price must be positive (example shows 0 and negatives are invalid) [cite: 1350, 1352]
            if price <= 0:
                print("Please enter a valid price!")
            else:
                return price
        except ValueError:
            # Handles non-numeric input like "pepperoni" [cite: 1351]
            print("Please enter a valid price!")

def main():
    print("Beckett Pizza Plaza 4-for-3 Offer")
    print("=================================") # [cite: 1344]

    prices = []
    
    # Collect 4 pizza prices [cite: 1339, 1340]
    for i in range(1, 5):
        price = get_valid_price(i)
        prices.append(price)

    # Calculate costs
    # The cheapest pizza is free [cite: 1339]
    cheapest_price = min(prices)
    total_original_cost = sum(prices)
    final_cost = total_original_cost - cheapest_price
    
    # Calculate discount percentage [cite: 1341]
    # Formula: (Amount Saved / Original Total) * 100
    if total_original_cost > 0:
        discount_percent = (cheapest_price / total_original_cost) * 100
    else:
        discount_percent = 0

    # Display results formatted to currency and integer percentage [cite: 1344, 1348]
    print(f"Order Total is £{final_cost:.2f}, a fabulous discount of {int(discount_percent)}%!")

if __name__ == "__main__":
    main()
```
